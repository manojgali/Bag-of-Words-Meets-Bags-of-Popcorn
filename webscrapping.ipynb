```
from bs4 import BeautifulSoup
import requests
import nltk
import html
from nltk.corpus import stopwords
import math
```

```
H = {}



depth = 0

urls = {}
sort={}
```

```
def crawl(u):
    r  = requests.get(u, "lxml")
    data = r.text
    soup = BeautifulSoup(data, "html.parser")
    return soup
```

```
def crawler(url,terminator=10):
    global urls
    global depth
    global sort
    global freqmatrix

    if depth < terminator:
        depth = depth+1
        crawlresult = crawl(url)
       
        links = crawlresult.find_all('a')

        urls[url]['out'] = len(links)

        #print(url + '\t is crawled')

        for link in links:
            tag = link.get('href')
            href_tag = str(tag)
            if href_tag[-1:] == '/':
                pass
            else:
                href_tag = href_tag + '/'
            if "http" in href_tag:
                if href_tag in urls.keys():
                    if 'in' in urls[href_tag].keys():
                        urls[href_tag]['in'] = urls[href_tag]['in']+1
                    elif 'out' in urls[href_tag].keys():
                        urls[href_tag]['in'] = 1
                    else:
                        urls[href_tag] = {}
                        urls[href_tag]['in'] = 1
                        crawler(href_tag)
                else:
                    urls[href_tag] = {}
                    urls[href_tag]['in'] = 1
                    crawler(href_tag)
```

```
start = 'http://www.google.co.in/'
urls[start] = {}

urls[start]['in'] = 1


crawler(start, terminator  = 50)
```

```
for url in urls.keys():
    if 'out' not in urls[url].keys():
        urls[url]['ratio'] = math.inf
    elif urls[url]['out'] == 0:
        urls[url]['ratio'] = -1
    else:
        urls[url]['ratio']= urls[url]['in']/urls[url]['out']
```

```
items = sorted(urls.items(), key = lambda tup: (tup[1]['ratio']))
for item in items:
    print(item[0], item[1]['ratio'])
```
